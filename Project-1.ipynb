3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
# Функция преобразующую опыт работы в месяцы
def get_expiriens(str):
    """Функция преобразующая признак опыта работы
    в опыт работы в месяцах

    Args:
        str (string): Строка из столбца "Опыт работы"

    Returns:
        int: Опыт работы в месяцах
    """
    if str is np.nan or str == 'Не указано':
        return np.nan
    else:
        experience_lst = str.split(maxsplit=6)
        experience = 0
        for i, elem in enumerate(experience_lst):
            if elem in ['года', 'год', 'лет']:
                experience += int(experience_lst[i - 1]) * 12
            if elem in ['месяцев', 'месяц', 'месяца']:
                experience += int(experience_lst[i - 1])

        return experience


# Новый признак ('Опыт работы (месяц)')
expiriens = hh_df['Опыт работы'].apply(get_expiriens)
hh_df['Опыт работы (месяц)'] = expiriens.round().astype('float16')

# Удаляю лишний признак ('Опыт работы')
hh_df = hh_df.drop('Опыт работы', axis=1)

# Чему равен медианный опыт работы (в месяцах) в нашей таблице?
median_exp = round(hh_df['Опыт работы (месяц)'].median())
print(f'Медианный опыт работы в данных: {median_exp} месяцев.')
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
# Задаю функцию определения города
def get_city(arg):
    """Функция извлекает город из строки и относит к одной из категорий:
    Москва, Санкт_петербург, город-миллионник, другие

    Args:
        arg (string): строка из признака "Город, переезд, командировки"

    Returns:
        string: строка, с наименованием категории
    """
    city = arg.split(',')[0].strip()
    million_cities = [
        'Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань',
        'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа',
        'Красноярск', 'Пермь', 'Воронеж', 'Волгоград'
    ]
    if city in million_cities:
        return 'город-миллионник'
    elif city in ['Москва', 'Санкт-Петербург']:
        return city
    else:
        return 'другие'


# Функция готовности к переезду
def ready_to_move(arg):
    """Функция, определяющая готов соискатель к переезду или нет

    Args:
        arg (string):строка из признака "Город, переезд, командировки"

    Returns:
        bool: True если готов к переезду, False в иных случаях
    """
    if 'перее' in arg:
        if ('не готова к перее' in arg) or ('не готов к перее' in arg):
            return False
        else:
            return True
    else:
        return False


# Функция готовности к командировкам
def ready_business_trip(arg):
    """Функция, определяющая готов соискатель к командировкам или нет

    Args:
        arg (string):строка из признака "Город, переезд, командировки"

    Returns:
        bool: True если готов к переезду, False в иных случаях
    """
    if 'командировк' in arg:
        if ('не готова к команд' in arg) or ('не готов к команд' in arg):
            return False
        else:
            return True
    else:
        return False


# Новый категориальный признак "Город"
city = hh_df['Город, переезд, командировки'].apply(get_city)
hh_df['Город'] = city.astype('category')

# Новый bool признак "Готовность к переезду"
to_move = hh_df['Город, переезд, командировки'].apply(ready_to_move)
hh_df['Готовность к переезду'] = to_move.astype('bool')

# Новый bool признак "Готовность к командировкам"
busi_trip = hh_df['Город, переезд, командировки'].apply(ready_business_trip)
hh_df['Готовность к командировкам'] = busi_trip.astype('bool')

# Удаляю лишний признак ('Город, переезд, командировки')
hh_df = hh_df.drop('Город, переезд, командировки', axis=1)

# Сколько процентов соискателей живут в Санкт-Петербурге?
city_percent = hh_df['Город'].value_counts(normalize=True)
spb_percent = round(city_percent.loc['Санкт-Петербург'] * 100)
print(f'В Санкт-Петербурге проживает {spb_percent}% соискателей.')

# Сколько процентов соискателей готовы одновременно и к переездам,
# и к командировкам?
mask_1 = hh_df['Готовность к переезду']
mask_2 = hh_df['Готовность к командировкам']
ready = round(hh_df[mask_1 & mask_2].shape[0] / hh_df.shape[0] * 100)
print(f'Одновременно к переезду и командировкам готовы {ready}% соискателя.')
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>

Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# Новые признаки занятости по маске(содержит ли строка необходимый признак)
employment = ['полная занятость', 'частичная занятость',
              'проектная работа', 'волонтерство', 'стажировка']
for elem in employment:
    hh_df[elem] = hh_df['Занятость'].str.contains(elem)

# Новые признаки графика
timing = ['полный день', 'сменный график', 'гибкий график',
          'удаленная работа', 'вахтовый метод']
for elem in timing:
    hh_df[elem] = hh_df['График'].str.contains(elem)

# Удаляю лишние признаки ('Занятость' и 'График')
hh_df = hh_df.drop(['Занятость', 'График'], axis=1)

# Сколько людей ищут проектную работу и волонтёрство?
print(f"Количество соискателей, которые ищут проектную работу и волонтерство:"
      f" {hh_df[hh_df['проектная работа'] & hh_df['волонтерство']].shape[0]}.")

# Сколько людей хотят работать вахтовым методом
# и с гибким графиком?
print(f"Количество людей, которые хотят работать вахтой, с гибким графиком:"
      f" {hh_df[hh_df['гибкий график'] & hh_df['вахтовый метод']].shape[0]}.")
```

Поэтому мы пойдем в лоб: обратимся к специальным интернет-ресурсам для получения данных о курсе в виде текстовых файлов. Например, MDF.RU, данный ресурс позволяет удобно экспортировать данные о курсах различных валют и акций за указанные периоды в виде csv файлов. Мы уже сделали выгрузку курсов валют, которые встречаются в наших данных за период с 29.12.2017 по 05.12.2019. Скачать ее вы можете **на платформе**

Создайте новый DataFrame из полученного файла. В полученной таблице нас будут интересовать столбцы:
* "currency" - наименование валюты в ISO кодировке,
* "date" - дата, 
* "proportion" - пропорция, 
* "close" - цена закрытия (последний зафиксированный курс валюты на указанный день).


Перед вами таблица соответствия наименований иностранных валют в наших данных и их общепринятых сокращений, которые представлены в нашем файле с курсами валют. Пропорция - это число, за сколько единиц валюты указан курс в таблице с курсами. Например, для казахстанского тенге курс на 20.08.2019 составляет 17.197 руб. за 100 тенге, тогда итоговый курс равен - 17.197 / 100 = 0.17197 руб за 1 тенге.

Осталось только понять, откуда брать дату, по которой определяется курс? А вот же она - в признаке **"Обновление резюме"**, в нем содержится дата и время, когда соискатель выложил текущий вариант своего резюме. Нас интересует только дата, по ней бы и будем сопоставлять курсы валют.

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
# DF с курсом валют
exchange_df = pd.read_csv(
    'data/ExchangeRates.csv',
    usecols=["currency", "date", "proportion", "close"]
)

# Перевожу даты в обеих таблицах в один формат
exchange_df['date'] = pd.to_datetime(
    exchange_df['date'], dayfirst=True, format="mixed"
).dt.date

hh_df['Обновление резюме'] = pd.to_datetime(
    hh_df['Обновление резюме'], dayfirst=True
).dt.date

# Выделяю из столбца ЗП валюту, заменяем ее в ISO формат
hh_df['Валюта'] = hh_df['ЗП'].apply(lambda x: x.split()[1])
hh_df['Валюта'] = hh_df['Валюта'].replace({
    'руб.': 'RUB', 'бел.руб.': 'BYN', 'грн.': 'UAH', 'сум': 'UZS'})

# Откидываю валюту из ЗП
hh_df['ЗП'] = hh_df['ЗП'].apply(lambda x: int(x.split()[0]))

# Присоединяю данные с курсом к таблице
hh_df = hh_df.merge(
    exchange_df,
    left_on=['Обновление резюме', 'Валюта'],
    right_on=['date', 'currency'],
    how='left'
)

# Заполняю пропущенные данные на единицу
hh_df['close'] = hh_df['close'].fillna(1)
hh_df['proportion'] = hh_df['proportion'].fillna(1)

# Вычисляю ЗП в рублях
hh_df['ЗП (руб)'] = (hh_df['ЗП'] * hh_df['close'] / hh_df['proportion'])
hh_df['ЗП (руб)'] = hh_df['ЗП (руб)'].astype('int32')

# Удаляю лишние признаки
extra_feature = ['ЗП', 'Валюта', 'currency', 'date', 'close', 'proportion']
hh_df = hh_df.drop(extra_feature, axis=1)

# Чему равна желаемая медианная заработная плата соискателей
# в нашей таблице (в рублях)?
salary_median = round(hh_df['ЗП (руб)'].median() / 1000)
print(f'Средний уровень желаемой зарплаты в данных: {salary_median} тыс.руб.')
```

Стоит отметить,по данному графику видно, что некоторые соискатели просто отмечают в анкете некоторые пункты совместно (или сразу все пункты), поэтому распределения и похожи между собой, имеет смысл проверить данные где True флаг установлен только в одном определенном виде занятости или графика работы:

```python
# Функция очищающая данные
def get_clean_data(data, check_list, parameter, new_graf, tr=3):
    """Функция очищающая таблицу от значений где выбрано более
    одной нужной нам категории (флаг True)

    Args:
        data (DataFrame): Изменяемая таблица
        check_list (list): список названий колонок, рассматриваемого признака
        parameter (string): рассматриваемый параметр (в нашем случае ЗП)
        new_graf (string): название нового признака
        tr (int, optional): Величина thresh в dropna. Defaults to 3.

    Returns:
        DataFrame: Возвращает таблицу, где удалены соискатели,
        у которых отмечено более одного флага True и есть новый признак с
        категорией этого флага
    """
    # Список признаков которые будут в новой таблице
    check_list.append(parameter)
    # Новая таблица с нужными признаками
    df = data.loc[:, check_list]
    # Избавляюсь от добавленного параметра из списка категорий
    check_list = check_list[:-1]
    # Временная копия таблицы
    temp_df = df.copy()
    # Из копии удалим признаки где заполнено только 2 значения
    # ('ЗП' и хотя бы 1 True)
    temp_df = temp_df.replace({False: np.nan}).dropna(thresh=tr)
    # В нужной таблице, по индексам из временной, сделаем подмену,
    # удаляем все признаки где больше 1 True
    df = df.drop(index=temp_df.index)
    # Добавим признак, в котором будет прописано единственное значение
    # категории
    df[new_graf] = ''
    for elem in check_list:
        df[elem] = df[elem].apply(
            lambda x: '<br>'.join(elem.split()).capitalize()
            if x is True else '')
        df[new_graf] += df[elem]
    return df


# Функция, возвращающую коробчатую диаграмму, нужного вида
def get_new_boxplot(data, cat, parameter):
    """Функция создающая коробчатую диаграмму,
    определенного вида

    Args:
        data (DataFrame): таблица из которой получаем данные
        cat (string): категориальный признак, который исследуется
        parameter (string): признак (числовой) распределение
        которого интересует

    Returns:
        figure: возвращает коробчатую диаграмму
    """
    fig = px.box(
       height=500,
       width=800,
       data_frame=data,
       x=parameter,
       y=cat,
       color=cat
    )
    fig.update_yaxes(
        # tickangle=330,
        gridcolor='silver',
        type='category',
        categoryorder='median descending'
    )
    fig.update_xaxes(gridcolor='silver', minor_griddash="dot")
    fig.update_layout(showlegend=False, boxgap=0, boxgroupgap=0.1)
    return fig
```

```python
def outliers_z_score(data, feature, log_scale=False, left=3, right=3):
    """Функция метода трех сигм

    Args:
        data (DataFrame): таблица с данными
        feature (Series): исследуемый признак
        log_scale (bool, optional): логарифмировать ли?. Defaults to False.
        left (int, optional): количество стандартных отклонений,
        для нижней границы. Defaults to 3.
        right (int, optional): количество стандартных отклонений,
        для верхней границы. Defaults to 3.

    Returns:
        DataFrame: возвращает 2 таблицы, первая с выбросами,
        вторая таблица отфильтрованные данные
    """
    if log_scale:
        x = np.log(data[feature])
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned


# Посмотрим на количество выбросов при отклонении на 3 сигмы от среднего
out_temp, hh_df_temp = outliers_z_score(df_log_age, 'Возраст', log_scale=True)
# Сколько выбросов вы нашли?
print(f'Количество найденных выбросов, при отклонении на '
      f'три сигмы в обе стороны: {out_temp.shape[0]}.')
age_out_temp = list(out_temp['Возраст'].value_counts().index)
print(f'Возраст соискателей, попавший под выбросы, при отклонении на 3 сигмы:'
      f'{[int(age) for age in age_out_temp]}.')

# Посмотрим на количество выбросов с послаблением в 1 сигму
outliers, hh_df = outliers_z_score(
    df_log_age,
    'Возраст',
    right=4,
    log_scale=True
)
age_out = list(outliers['Возраст'].value_counts().index)
# Сколько выбросов вы нашли?
print(f'Количество найденных выбросов, при послаблении на 1 сигму: '
      f'{outliers.shape[0]}.')
print(f'Возраст соискателей, попавший под выбросы:'
      f'{[int(age) for age in age_out]}.')
# Вернем признак возраста в нужный нам формат
hh_df.loc[:, 'Возраст'] = hh_df.loc[:, 'Возраст'].astype('int8')
# Итоговое количество строк в таблице
print(f'Всего в таблице осталось: {hh_df.shape[0]} строк.')
```
